## 2. Trimming

### 2.1a Trim_galore

```bash
#!/bin/bash

PROJECT="/workspace/cfngle/raw-data/JM"
IN="${PROJECT}/raw-reads"
OUT="${PROJECT}/002.trim_galore"

mkdir -p $IN
mkdir -p $OUT


# Get the files to trim:
FILES=`basename -a ${IN}/*.fastq`

echo $FILES

module load trim_galore

for FILE in $FILES
     do
        In_File1=${IN}/${FILE}
        echo $In_File1
        Out=${OUT}/${FILE}_trimmed.fastq
        COMMAND="trim_galore --rrbs --quality 25 --clip_R1 4 --three_prime_clip_R1 2 $In_File1 --output_dir $OUT"
        
        echo $COMMAND
done | abatch -j ${OUT}/trim_galore-logs --time 04:00:00 --mem 1G | sbatch
```

### 2.1b Trimmomatic

```bash
#!/bin/bash

PROJECT="/workspace/cfngle/raw-data/JM"
IN="${PROJECT}/raw-reads"
OUT="${PROJECT}/002.trimmomatic/"
ADAPTER="${PROJECT}/zzz.adapters/TruSeq3-SE.fa"

mkdir -p $IN
mkdir -p $OUT


# Get the files to trim:
FILES=`basename -a ${IN}/*.fastq`

#echo $FILES

module load Trimmomatic

for FILE in $FILES
     do
        In_File1=${IN}/${FILE}
        echo $In_File1
        Out=${OUT}${FILE}_trimmoniac.fastq
        COMMAND="trimmomatic SE -phred33 ${In_File1} ${Out} ILLUMINACLIP:${ADAPTER}:2:30:10 LEADING:3 TRAILING:3 SLIDINGWINDOW:4:15 MINLEN:36"
        
        echo $COMMAND
done | abatch -j ${OUT}/trimmomatic-logs --time 04:00:00 --mem 1G | sbatch
```

### 5.2 Alignments to genome

```bash
#!/bin/bash

PROJECT="/workspace/cfngle/raw-data/JM"
IN="${PROJECT}/002.trim_galore"
OUT="${PROJECT}/004.alignments"
GENOME="/workspace/cfngle/raw-data/JM/rgenome"
LOG="${OUT}/logs"

mkdir -p $LOG

module load bismark/0.23.0

# Use echo statements to be sure that the results from awk are what you really want...
# FILES=`ls ${IN}/*.fq | awk -F'[_ ]' '{print $1"_"$2"_"$3"_"$4"_"$5}' | sort -u`
FILES=`ls ${IN}/*.fq`
# echo $FILES

for file in ${FILES}
    do
        NAME=`basename ${file}`
        PREFIX=`echo ${NAME} | awk -F[_] '{print $1"_"$2"_Alignments"}'`
        #echo ${PREFIX}
        #echo ${file}
        COMMAND="bismark \
                --genome ${GENOME} \
                $file \
                --o ${OUT}/${PREFIX} \
                --multicore 5 \
                --non_directional"
       # echo "${OUT}${PREFIX}"
       echo $COMMAND

done | abatch -j ${OUT}/bis-align-local --time 23:00:00 --mem 25G -c 1 | sbatch
```

### 7. Methylation extraction

```bash
#!/bin/bash

PROJECT="/workspace/cfngle/raw-data/JM"
IN="${PROJECT}/004.alignments"
OUT="${PROJECT}/007.meth_extraction"
GENOME="/workspace/cfngle/raw-data/JM/rgenome"

module load bismark

FILES=`find ${IN}/ -type f -name "*.bam"`
#echo ${FILES}

for file in $FILES
    do
        COMMAND="bismark_methylation_extractor \
                ${file} \
                -s \
                --merge_non_CpG \
                --cytosine_report \
                --scaffolds \
                --genome_folder ${GENOME} \
                --multicore 6 \
                -o ${OUT}"
      echo $COMMAND

done | abatch -j ${OUT}/meth-extract --time 5-01:00:00 --mem 4G -c 1| sbatch
```
