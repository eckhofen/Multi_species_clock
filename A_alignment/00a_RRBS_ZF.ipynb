### 1.1 FastQC RAW Data
- The input for this step is the raw data from the provider in FASTQ format
- The output from this step are the HTML FASTQC Reports

```bash
#!/bin/bash

PROJECT="/workspace/cfngle/raw-data/ZF/raw-reads" 
IN="${PROJECT}/"
OUT="${PROJECT}/001.fastqc_raw"

cd $IN
#  FILES=`find -type f -name "*.fastq"`
FILES=`find -type f -name "*.fastq"`
echo $FILES


# Load the FastQC module:
module load FastQC

for file in $FILES
    do
         COMMAND="fastqc --nogroup -q -t 2 -o ${OUT} ${file}"
         echo $COMMAND
        #echo $file
done | abatch -j ${OUT}/fastqc_raw --time 01:00:00 --mem 20G -c 10 | sbatch
```

## 2. Trimming

### 2.1a Trim_galore

```bash
#!/bin/bash

PROJECT="/workspace/cfngle/raw-data/ZF"
IN="${PROJECT}/raw-reads"
OUT="${PROJECT}/002.trim_galore"

mkdir -p $IN
mkdir -p $OUT


# Get the files to trim:
FILES=`basename -a ${IN}/*.fastq`

echo $FILES

module load trim_galore

for FILE in $FILES
     do
        In_File1=${IN}/${FILE}
        echo $In_File1
        Out=${OUT}/${FILE}_trimmed.fastq
        COMMAND="trim_galore --rrbs --quality 25 --clip_R1 4 --three_prime_clip_R1 2 $In_File1 --output_dir $OUT"
        
        echo $COMMAND
done | abatch -j ${OUT}/trim_galore-logs --time 04:00:00 --mem 1G | sbatch
```

### 2.1b Trimmomatic

```bash
#!/bin/bash

PROJECT="/workspace/cfngle/raw-data/ZF"
IN="${PROJECT}/raw-reads"
OUT="${PROJECT}/002.trimmomatic/"
ADAPTER="${PROJECT}/zzz.adapters/TruSeq3-SE.fa"

mkdir -p $IN
mkdir -p $OUT


# Get the files to trim:
FILES=`basename -a ${IN}/*.fastq`

#echo $FILES

module load Trimmomatic

for FILE in $FILES
     do
        In_File1=${IN}/${FILE}
        echo $In_File1
        Out=${OUT}${FILE}_trimmoniac.fastq
        COMMAND="trimmomatic SE -phred33 ${In_File1} ${Out} ILLUMINACLIP:${ADAPTER}:2:30:10 LEADING:3 TRAILING:3 SLIDINGWINDOW:4:15 MINLEN:36"
        
        echo $COMMAND
done | abatch -j ${OUT}/trimmomatic-logs --time 04:00:00 --mem 1G | sbatch
```

```bash
#!/bin/bash

PROJECT="/workspace/cfngle/raw-data/ZF"
IN="${PROJECT}/004.alignments"
OUT="${PROJECT}/007.meth_extraction"
GENOME="/workspace/cfngle/raw-data/ZF/rgenome"

module load bismark

FILES=`find ${IN}/ -type f -name "*.bam"`
#echo ${FILES}

for file in $FILES
    do
    #echo ${file}
        COMMAND="bismark_methylation_extractor \
                ${file} \
                -s \
                --merge_non_CpG \
                --cytosine_report \
                --scaffolds \
                --genome_folder ${GENOME} \
                -o ${OUT}"
      echo $COMMAND

done | abatch -j ${OUT}/meth-extract --time 5-01:00:00 --mem 5G | sbatch
```
