```bash
##03 index (.bam)
cd "${path_results}bowtie2/"

for files in sorted*; do
    # Index
    samtools index "$files"

    echo "indexed $files"
done
```

```bash
## sort (.bam)
cd "${path_results}bowtie2/"

for bamfile in *.bam; do
    # Output is defined as bam file + "sorted"
    bamfile_sorted="sorted_$bamfile"

    # Sort bam files
    samtools sort "$bamfile" > "$bamfile_sorted"

    echo "Sorted $bamfile to $bamfile_sorted"
done
```

```bash
## converting .sam to .bam

cd "${path_results}bowtie2/"

# Loop through all .sam files in the directory
for samfile in *.sam; do
    # Define the output BAM file name by replacing the .sam extension with .bam
    bamfile="${samfile%.sam}.bam"

    # Convert SAM to BAM
    samtools view -Sb "$samfile" > "$bamfile"

    echo "Converted $samfile to $bamfile"
done
```
